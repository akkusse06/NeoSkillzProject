```python
# Ensure consistency in categorical columns
# List of columns that need cleaning (strip + lowercase)
columns_to_clean = ['CLIENT_ID', 'PERSON_HOME_OWNERSHIP', 'LOAN_INTENT','LOAN_GRADE','CB_PERSON_DEFAULT_ON_FILE','GENDER','MARITAL_STATUS',
                    'EDUCATION_LEVEL','COUNTRY','STATE','CITY','EMPLOYMENT_TYPE','LOAN_STATUS_DESC']

# For loop version
for col in columns_to_clean:
    df[col] = df[col].str.strip().str.lower()
```

```python
df.drop(columns = ['CITY_LATITUDE','CITY_LONGITUDE','CLIENT_ID','LOAN_STATUS_DESC'],inplace = True)
```

```python
df.columns
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
numeric_cols = [
    'PERSON_AGE', 'PERSON_INCOME', 'PERSON_EMP_LENGTH',
    'LOAN_AMNT', 'LOAN_INT_RATE', 'LOAN_PERCENT_INCOME',
    'CB_PERSON_CRED_HIST_LENGTH', 'LOAN_TERM_MONTHS',
    'LOAN_TO_INCOME_RATIO', 'OTHER_DEBT', 'DEBT_TO_INCOME_RATIO',
    'OPEN_ACCOUNTS', 'CREDIT_UTILIZATION_RATIO', 'PAST_DELINQUENCIES'
]
```

```python
scaler = StandardScaler()
```

```python
categorical_cols = [
    'PERSON_HOME_OWNERSHIP', 'LOAN_INTENT', 'LOAN_GRADE',
    'CB_PERSON_DEFAULT_ON_FILE', 'GENDER', 'MARITAL_STATUS',
    'EDUCATION_LEVEL', 'EMPLOYMENT_TYPE','COUNTRY', 'STATE', 'CITY'
]
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)
```

```python
X = df.drop(columns='LOAN_STATUS')
```

```python
y = df[['LOAN_STATUS']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print(f"Training samples: {len(X_train):,}")
print(f"Test samples:     {len(X_test):,}\n")
```

```python
X_train_num_scaled = scaler.fit_transform(X_train[numeric_cols])
X_test_num_scaled  = scaler.transform(X_test[numeric_cols])
```

```python
X_train_cat_encoded = encoder.fit_transform(X_train[categorical_cols])
X_test_cat_encoded  = encoder.transform(X_test[categorical_cols])
```

```python
X_train_final = np.hstack([X_train_num_scaled, X_train_cat_encoded])
X_test_final  = np.hstack([X_test_num_scaled,  X_test_cat_encoded])
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train_final,y_train)
```

```python
y_pred = log.predict(X_test_final)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

# LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_final,y_train)
```

```python
y_pred_linear = model.predict(X_test_final)
```

```python
# Convert regression output → binary class 
y_pred_class = (y_pred_linear > 0.5).astype(int)   # the threshold at 0.5
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
mae = mean_absolute_error(y_test,y_pred_linear)
mae
```

```python
mse = mean_squared_error(y_test,y_pred_linear)
mse
```

```python
R2 = r2_score(y_test,y_pred_linear)
R2
```

```python
RMSE = root_mean_squared_error(y_test,y_pred_linear)
RMSE
```

```python
acc_linear = accuracy_score(y_test, y_pred_class)
acc_linear
```

```python
cm_linear = confusion_matrix(y_test, y_pred_class)
cm_linear
```
